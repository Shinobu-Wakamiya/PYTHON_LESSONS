```python
import numpy as np
import matplotlib.pyplot as plt
import random
import math
import seaborn as sns
```

Generate random number and Obtain X & Visualised Y and X distribution

```python
# Define the value of 'a'
a = 1.5  # example value, set 'a' as needed

# Initialize a list to store the values of X
X_values = []
Y_val = []

# Iterate 100 times
for _ in range(10000):
    Y = random.random()
    X = math.exp(-3.5065) * ((1 - Y) ** (-1 / a))
    X_values.append(X)
    Y_val.append(Y)

# Plot the distribution of X
sns.histplot(X_values, kde=True, stat="density", bins=100)
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Probability Distribution')
plt.show()



sns.histplot(Y_val, kde=True, stat="density", bins=100)
plt.xlabel('Y')
plt.ylabel('Probability Density')
plt.title('Probability Distribution')
plt.show()

```

STEP 1: Generate list of Resetting time drawn from exponential distribution: List is given by T_values

```python
# Define the value of 'r' = 1.5  # example value, set 'a' as needed
ro=0.001
To=0
# Initialize a list to store the values of X
T_values = []
R_val = []

# Iterate 100 times
for _ in range(10000):
    R = random.random()
    T = (np.log(math.exp(ro*To)/(1-(R*math.exp(ro*To)))))/ro
    T_values.append(T)
    R_val.append(R)

# Plot the distribution of X
sns.histplot(T_values, kde=True, stat="density", bins=100)
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Probability Distribution')
plt.show()



sns.histplot(R_val, kde=True, stat="density", bins=100)
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Probability Distribution')
plt.show()
```

```python
def random_walk(step_limit, reset_time_list, p, num_realizations):
    all_displacements = []
    all_total_times = []
    long = []
    
    for _ in range(num_realizations):
        position = 0
        total_time = 0
        displacement = []
        total_time_array = []

        while total_time < step_limit:
            reset_time = int(np.random.choice(reset_time_list))

            for t in range(reset_time):
                move = np.random.choice([-1,1],p=[p,(1-p)])
                position += abs(move)
                displacement.append(position)
                total_time_array.append(total_time+t+1)
            
                if total_time+t+1 >= step_limit:
                    break

            total_time += reset_time
            long.append(position)
            position = 0
            

        all_displacements.append(displacement)
        all_total_times.append(total_time_array)

#    all_displacements = np.array(all_displacements)
#    all_total_times = np.array(all_total_times)

    avg_displacement = np.mean(all_displacements, axis=0)
    avg_total_time = np.mean(all_total_times, axis=0)

    return avg_total_time, avg_displacement, long

step_limit = 5000
reset_time_list = T_values
num_realizations = 5000
p_values_list = X_values

#Lists to store long time displacements fro two diffrent p values
long_time_displacements_p1 = {}

#plotting setup
plt.figure(figsize=(10,6))

pi=[]
qi=[]

for i in range(10):
    p = np.random.choice(p_values_list)
    while p>=1:
        p = np.random.choice(p_values_list)
    q=1-p
    pi.append(p)
    qi.append(q)


    avg_total_time, avg_displacement, long = random_walk(step_limit, reset_time_list, p, num_realizations)

    long_time_displacements_p1[p] = long
    
    plt.plot(avg_total_time, avg_displacement, label=f'p={p:.2f}')
    



plt.plot(avg_total_time, avg_displacement)
plt.xlabel('Total Time (Steps)')
plt.ylabel('Average Displacement')
plt.title('Total Time Vs Average Dis')
plt.legend()
plt.savefig('mean_displacement_new1.png',dpi=300)
plt.show()

for p in long_time_displacements_p1:
    print(f'long time displacements for p ={p:.2f}: {long_time_displacements_p1[p][:10]}')
```

```python
step_limit = 500
reset_time_list = T_values
num_realizations = 500
p_values_list = X_values
num_sites = 10000



def random_walk(step_limit, reset_time_list, p, num_realizations):
    all_displacements = []
    all_total_times = []
    long = []
    
    for _ in range(num_realizations):
        position = num_sites // 2
        total_time = 0
        displacement = []
        total_time_array = []
        

        while total_time < step_limit:
            reset_time = int(np.random.choice(reset_time_list))

            for t in range(reset_time):
                pe = np.random.choice(X_values)
                qe = 1-pe
                move = np.random.choice([-1,1],p=[pe,qe])
                position += move

                #Boundary Condition
                if position < 0:
                    position = num_sites - 1
                elif position >= num_sites:
                    position = 0
                    
                displacement.append(abs(position))
                total_time_array.append(total_time+t+1)
            
                if total_time+t+1 >= step_limit:
                    break

            total_time += reset_time
            long.append(position)
            position = num_sites // 2
            

        all_displacements.append(displacement)
        all_total_times.append(total_time_array)

#    all_displacements = np.array(all_displacements)
#    all_total_times = np.array(all_total_times)

    avg_displacement = np.mean(all_displacements, axis=0)
    avg_total_time = np.mean(all_total_times, axis=0)

    return avg_total_time, avg_displacement, long



#Lists to store long time displacements fro two diffrent p values
long_time_displacements_p1 = {}

#plotting setup
plt.figure(figsize=(10,6))

pi=[]
qi=[]

for i in range(10):
    p = np.random.choice(p_values_list)
    while p>=1:
        p = np.random.choice(p_values_list)
    q=1-p
    pi.append(p)
    qi.append(q)


    avg_total_time, avg_displacement, long = random_walk(step_limit, reset_time_list, p, num_realizations)

    long_time_displacements_p1[p] = long
    
    plt.plot(avg_total_time, avg_displacement, label=f'p={p:.2f}')
    



plt.plot(avg_total_time, avg_displacement)
plt.xlabel('Total Time (Steps)')
plt.ylabel('Average Displacement')
plt.title('Total Time Vs Average Dis')
plt.legend()
plt.savefig('mean_displacement_new1.png',dpi=300)
plt.show()

for p in long_time_displacements_p1:
    print(f'long time displacements for p ={p:.2f}: {long_time_displacements_p1[p][:10]}')
```
